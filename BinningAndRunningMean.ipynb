```python
def get_cryosat_freeboard(mib, mplot, year, month, day, ridge_stats_path):

        month1 = '%02d' %(month)
        day1 = '%02d' %(day)
        #ridge_stats_path='/Volumes/arco/'
        print (ridge_stats_path)
        files = glob.glob(ridge_stats_path+str(year)+str(month1)+'_MERGE/cry_NO_*'+str(year)+month1+day1+'*.fb')
        if (np.size(files)==0):
                print ('No data this year')
        lats=[]
        lons=[]
        freeboard=[1]
        concentration=[]
        sea_ice_type=[]
        flag=[]
        flag_snow=[]

        for i in range(np.size(files)):
                print (i, np.size(files))
                data = np.loadtxt(files[i])
                if (data.shape[0]>=1):
                        lats.extend(data[:,3])
                        lons.extend(data[:,4])
                        freeboard.extend(data[:,5])
                        concentration.extend(data[:,8])
                        sea_ice_type.extend(data[:,9])
                        flag.extend(data[:,10])
                        flag_snow.extend(data[:,11])

#        lons, lats = mib(xpts, ypts, inverse=True)
        xpts, ypts = mplot(lons, lats)
#        return np.array(xpts), np.array(ypts), np.array(lats), np.array(lons), np.array(freeboard), np.array(concentration), np.array(sea_ice_type), np.array(flag), np.array(flag_snow)
        return np.array(xpts), np.array(ypts), np.array(lats), np.array(lons), np.array(freeboard), np.array(concentration), np.array(sea_ice_type), np.array(flag)#, np.array(flag_snow)

    
```

```python
def get_cryosat_thickness(mib, mplot, year, month, day, ridge_stats_path):

        month1 = '%02d' %(month)
        day1 = '%02d' %(day)
        #ridge_stats_path='/Volumes/arco/'
        print (ridge_stats_path)
        files = glob.glob(ridge_stats_path+str(year)+str(month1)+'_MERGE/cry_NO_*'+str(year)+month1+day1+'*.thk')
        if (np.size(files)==0):
                print ('No data this year')
        lats=[]
        lons=[]
        validity=[]
        thickness=[]
        concentration=[]
        sea_ice_type=[]

        for i in range(np.size(files)):
                print (i, np.size(files))
                data = np.loadtxt(files[i])
                if len(data.shape)>1:
                        if (data.shape[0]>=1):
                                lats.extend(data[:,3])
                                lons.extend(data[:,4])
                                validity.extend(data[:,5])
                                thickness.extend(data[:,6])
                                sea_ice_type.extend(data[:,7])
                                concentration.extend(data[:,9])
                else:
                        if (data.shape[0]>=1):
                                lats.append(data[3])
                                lons.append(data[4])
                                validity.append(data[5])
                                thickness.append(data[6])
                                sea_ice_type.append(data[7])
                                concentration.append(data[9])



#        lons, lats = mib(xpts, ypts, inverse=True)
        xpts, ypts = mplot(lons, lats)
        return np.array(xpts), np.array(ypts), np.array(lats), np.array(lons), np.array(thickness), np.array(sea_ice_type), np.array(concentration), np.array(validity)
```

```python
def get_cryosat_elevation(mib, mplot, year, month, day, ridge_stats_path):

        month1 = '%02d' %(month)
        day1 = '%02d' %(day)
        #ridge_stats_path='/Volumes/arco/'
        print (ridge_stats_path)
        files = glob.glob(ridge_stats_path+str(year)+str(month1)+'_MERGE/cry_NO_*'+str(year)+month1+day1+'*.elev')
        if (np.size(files)==0):
                print ('No data this year')
        surface_type=[]
        validity=[]
        lats=[]
        lons=[]
        time=[]
        concentration=[]
        sea_ice_type=[]
        elev_abs=[]
        elev_mean=[]

        for i in range(np.size(files)):
                print (i, np.size(files))
                data = np.loadtxt(files[i])
                if (data.shape[0]>=1):
                        surface_type.extend(data[:,0])
                        validity.extend(data[:,1])
                        lats.extend(data[:,5])
                        lons.extend(data[:,6])
                        time.extend(data[:,4])
                        concentration.extend(data[:,11])
                        sea_ice_type.extend(data[:,12])
                        elev_abs.extend(data[:,7])
                        elev_mean.extend(data[:,8])

#        lons, lats = mib(xpts, ypts, inverse=True)
        xpts, ypts = mplot(lons, lats)
        return np.array(xpts), np.array(ypts), np.array(lats), np.array(lons), np.array(elev_abs), np.array(elev_mean), np.array(concentration), np.array(sea_ice_type), np.array(surface_type), np.array(validity), np.array(time)
```
